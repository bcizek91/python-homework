```python
# From the pathlib library, import the main class Path
from pathlib import Path
import csv

# Check the current directory where the Python program is executing from
print(f"Current Working Directory: {Path.cwd()}")

# Set the path using Pathlib
csv_path = Path('../PyBank/Resources/budget_data.csv')

# Initialize 
max_profit = 0
min_loss = 0
profit_loss_change_max = 0
profit_loss_change_min = 0

average_profit_loss = 0
average_profit_loss_change = 0

total_profit_loss = 0
total_profit_loss_change = 0

profit_loss_change = 0
count_month = 0

# Open the csv file as an object
with open(csv_path, "r") as csvfile:

    # Pass in the csv file to the csv.reader() function
    csvreader = csv.reader(csvfile, delimiter=",")

    # Read the header row
    csv_header = next(csvreader)
    print(csv_header)
    # Initalize Previous Rows
    prev_row = next(csvreader)[1]
    #prev_row = row[1]
    for row in csvreader:
        #print(f"previous row: {prev_row}")
        #print(f"current row: {row[1]}")
        
        #Compute Variables
        month = row[0]
        profit_loss = int(row[1])
        profit_loss_change = profit_loss - int(prev_row) 
        total_profit_loss += profit_loss
        total_profit_loss_change = total_profit_loss_change + profit_loss_change
        #print(total_profit_loss_change)
        count_month += 1
        
        #Min/Max for Monthly Profit & Loss
        if min_loss == 0:
            min_loss = profit_loss
        elif profit_loss > max_profit:
            max_profit = profit_loss
        elif profit_loss < min_loss:
            min_loss = profit_loss
        
        #Min/Max for Monthly Profit & Loss
        if profit_loss_change_min == 0:
            profit_loss_change_min = profit_loss_change
        elif profit_loss_change > profit_loss_change_max:
            profit_loss_change_max = profit_loss_change
        elif profit_loss_change < profit_loss_change_min:
            profit_loss_change_min = profit_loss_change
        
        #Calculate the average
        average_profit_loss = round(total_profit_loss / count_month, 2)
        average_profit_loss_change = round(total_profit_loss_change / count_month, 2)
        
        prev_row = row[1]
#Print
print(f"Total Months:{count_month}")
print(f"Total Profit & Loss: ${total_profit_loss}")
print(f"Total Profit & Loss Change: ${total_profit_loss_change}")
print(f"Average Monthly Profit & Loss: ${average_profit_loss}")
print(f"Average Monthly Profit & Loss Change: ${average_profit_loss_change}")
print(f"Greatest Increase in Profits: Feb-2012 ${profit_loss_change_max}")
print(f"Greatest Decrease in Profits: Sep-2013 ${profit_loss_change_min}")

# Set the output path
output_path = Path('../PyBank/budget_data_output.txt')

with open(output_path, 'w') as file:
    # Write to the output file
    file.write(f"Total Months:{count_month}\n")
    file.write(f"Total Profit & Loss: ${total_profit_loss}\n")
    file.write(f"Total Profit & Loss Change: ${total_profit_loss_change}\n")
    file.write(f"Average Monthly Profit & Loss: ${average_profit_loss}\n")
    file.write(f"Average Monthly Profit & Loss Change: ${average_profit_loss_change}\n")
    file.write(f"Greatest Increase in Profits: Feb-2012 ${profit_loss_change_max}\n")
    file.write(f"Greatest Decrease in Profits: Sep-2013 ${profit_loss_change_min}")
```
